```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

def sigmoid_activation(x):
  return (1/(1+np.exp(-x)))

def derivative_sigmoid(y):
  return ((sigmoid_activation(y)*(1-sigmoid_activation(y))))

def error(target,calculated):
  return (calculated-target)

def relu(x):
  return np.maximum(0,x)

def derivative_relu(x):
  # Create a copy of the input array to avoid modifying the original
  result = np.copy(x)
  result[result<=0] = 0
  result[result>0] = 1
  return result # Return the modified copy

W1=np.array([[random.random(),random.random()],[random.random(),random.random()]])
W2=np.array([[random.random()],[random.random()]])
biasno=random.random()
b1=np.array([[biasno],[biasno]])

X=np.array([[0.8],[0.1]])

target=np.array([[0.9]])
changeinW2=np.zeros((2,1))
print(changeinW2)
changeinW1=np.zeros((2,2))
#W2_new=np.zeros((2,2))
#W1_new=np.zeros((2,2))
error_list=[]
xpoints_list=[]
for k in range (0,2): #20085
  H1_in=(np.dot(W1,X))+b1
  H1_out=relu(H1_in)
  H1_derivative=derivative_relu(H1_in)

  H2_in=(H1_out[0][0]*W2[0][0]+H1_out[1][0]*W2[1][0])
  H2_out=relu(H2_in)
  # Pass a copy of H2_in to derivative_relu to avoid modifying the original
  H2_derivative=derivative_relu(np.copy(H2_in))

  TotalError=(1/2)*((error(1,H2_out))**2)


  error_list.append(TotalError)
  xpoints_list.append(k)
  for i in range(0,2):

    h_out=H1_out[i][0]
    derivative = H2_derivative
    errors=error(target,H2_out)
    changeinW2[i][0]=h_out*derivative*errors
  # changeinW2=changeinW2.transpose()


  W2=W2-(0.5*changeinW2) #learning rate=0.5

  for i in range(0,2):
    for j in range(0,2):
      input=X[i][0]
      if (j==0):
        sum=(W2[0][0]*H2_derivative*error(target,H2_out))
      else:
        sum=(W2[1][0]*H2_derivative*error(target,H2_out))
      changeinW1[i][j]=input*H1_derivative[j][0]*sum
    #changeinW1[i][j]=input*H1_derivative[j][0]*((W2[i][j]*H2_derivative[j][0]*error(target[j][0],H2_out[j][0])) + (W2[i][j]*H2_derivative[j][0]*error(target[j][0],H2_out[j][0])))
    changeinW1=changeinW1.transpose()


  W1=W1-(0.5*changeinW1)

  changeinB1=np.zeros((2,1))
  changeinB1[0][0]=(error(target,H2_out)*H2_derivative*(W2[0][0]*H1_derivative[0][0]+W2[1][0]*H1_derivative[1][0]))
  changeinB1[1][0]=(error(target,H2_out)*H2_derivative*(W2[0][0]*H1_derivative[0][0]+W2[1][0]*H1_derivative[1][0]))
  #print(changeinB1)
  b1=b1-(0.5*changeinB1)
  #print(b1) #new b1
# W1=W1_new
# W2=W2_new
print(W1)
print(W2)
print(H2_out)
print(error_list[-1])
xpoints = np.array(xpoints_list)
ypoints = np.array(error_list)

plt.plot(xpoints, ypoints)
plt.show()
```
